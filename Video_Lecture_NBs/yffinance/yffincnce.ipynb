```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ticker = ["AAPL","BA","KO","IBM","DIS","MSFT"]
```

```python
stocks = yf.download(ticker, start = "2010-01-01", end = "2023-03-20")
```

```python
stocks.info()
```

```python
stocks.to_csv("stocks.csv")
```
